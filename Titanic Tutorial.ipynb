## Random Forest Approach

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
x = pd.get_dummies(train_data[features])
x_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(x, y)
predictions = model.predict(x_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('dataset/submission.csv', index=False)
print("Saved")
```

## Neural Network using Keras

```python
# libs
import pandas as pd
import numpy as np

data_train = pd.read_csv('dataset/train.csv')
data_test = pd.read_csv('dataset/test.csv')
```

```python
# data_check = pd.read_csv('dataset/data_check.csv')
```

```python
# remove unused columns
# data_train = data_train.drop(['Name', 'Ticket', 'Fare', 'Cabin', 'Embarked', 'Age'], axis=1)
# data_test = data_test.drop(['Name', 'Age', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1)
# data_train.head()
data_test.head()
```

```python
# Import LabelEncoder from Sklearn
from sklearn.preprocessing import LabelEncoder
label_encoder_sex = LabelEncoder()
```

```python
# Transform sex column values using label Encoder

data_train.iloc[:,1] = label_encoder_sex.fit_transform(data_train.iloc[:,1])
```

```python
data_test.iloc[:,1] = label_encoder_sex.fit_transform(data_test.iloc[:,1])
```

```python
data_train = data_train[['PassengerId', 'Sex', 'SibSp', 'Parch', 'Pclass', 'Survived']]
```

```python
data_test = data_test[['PassengerId', 'Sex', 'SibSp', 'Parch', 'Pclass']]
```

```python
data_test.head()
```

```python
data_train.head()
```

```python
X_train = data_train.iloc[:,0:5] #Inputs
y_train = data_train.iloc[:,5] #Output (Survived)
```

```python
# normalise data
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(data_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
#Input Layer with 5 input neutrons

classifier.add(Dense(3, kernel_initializer = 'uniform', activation = 'relu', input_shape = (5,)))

#Hidden Layer

classifier.add(Dense(2, kernel_initializer = 'uniform', activation = 'relu'))

#output layer with 1output neuron which will predit 1 or 0

classifier.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# getting predictions of test data
prediction = classifier.predict(X_test).tolist()

# list to series
se = pd.Series(prediction)

# creating new column of predictions in data_check dataframe
data_check = pd.read_csv('dataset/gender_submission.csv')
data_check['check'] = se
data_check['check'] = data_check['check'].str.get(0)
```

```python
# check over 0,5 = predicted 1
data_check.head()
```

```python
series = []
for val in data_check.check:
    if val >= 0.5:
        series.append(1)
    else:
        series.append(0)
        
data_check['final'] = series
```

```python
match = 0
nomatch = 0
for val in data_check.values:
    if val[1] == val[3]:
        match = match +1
    else:
        nomatch = nomatch +1
```

```python
print("match:", match)
print("nomatch:", nomatch)
```

```python
print((match / (match+nomatch))*100, "%")
```
